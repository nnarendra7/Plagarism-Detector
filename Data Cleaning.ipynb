```python
import pyarrow.parquet as pq
import json
import pandas as pd
df = pq.read_table(source="data.pq").to_pandas()
df
```

```python
content = None
for i,val in enumerate(df['content']):
    content = pd.concat([content,pd.json_normalize(json.loads(val))])
content.head(1)
```

```python
content_source = None
for i,val in enumerate(df['content_source']):
    content_source = pd.concat([content_source,pd.json_normalize(json.loads(val))])
content_source.head(1)
```

```python
import pandas as pd
search = None
for i,val in enumerate(df['search']):
    search = pd.concat([search,pd.json_normalize(json.loads(val))])
search.head(1)
```

```python
course = None
for i,val in enumerate(df['course']):
    course = pd.concat([course,pd.json_normalize(json.loads(val))])
course.head(1)
# course['data.etc.coursehero.num_documents'].value_counts()
```

```python
sub_organization = None
for i,val in enumerate(df['sub_organization']):
    sub_organization = pd.concat([sub_organization,pd.json_normalize(json.loads(val))])
sub_organization.head(1)
```

```python
organization = None
for i,val in enumerate(df['organization']):
    organization = pd.concat([organization,pd.json_normalize(json.loads(val))])
organization.head(1)
```

```python
organizations= pd.concat([course,sub_organization,organization],axis=1)
contents = pd.concat([content,content_source,search],axis=1)
```

```python
orgs = organizations.T.drop_duplicates().T
orgs.head(2)
```

```python
cont = contents.loc[:,~contents.columns.duplicated()]
cont.head(2)
```

```python
con_orgs = pd.concat([cont,orgs],axis=1)
```

```python
final = con_orgs.loc[:,~con_orgs.columns.duplicated()]
final.head(2)
```

```python
temp = df.drop(['content','content_source','search','course','sub_organization','organization'],axis=1)
```

```python
final.reset_index(inplace=True, drop=True)
df.reset_index(inplace=True, drop=True)
temp.reset_index(inplace=True, drop=True)
```

```python
data = pd.concat([temp,final],axis=1)
```
